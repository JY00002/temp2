```python
#labels=['8PSK','AM-DSB','BPSK','CPFSK','GFSK','PAM4','QAM16','QAM64','QPSK','WBFM']
#labels=['8PSK','AM-DSB-WC','BPSK','16APSK','GMSK','4ASK','16QAM','64QAM','QPSK','FM']
labels = ['BPSK','QPSK','8PSK','GMSK','16APSK','64APSK','16QAM','64QAM','AM-DSB-WC','FM']


import pickle
conmat=pickle.load(open('Confusion matrix_Model_C_90classes_Data_norm_DOST_no_norm_concatenated_4x128.dat','rb'))

import matplotlib.pyplot as plt
import numpy as np

conmat.keys()

def plot_confusion_matrix(cm, title='Confusion matrix', cmap=plt.cm.Blues, labels=[]):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(labels))
    plt.xticks(tick_marks, labels, rotation=45)
    plt.yticks(tick_marks, labels)
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
k=['all',-8,-6,-4,-2,0,2,4,6,8]
for keys in k:
    acc=np.sum(np.diagonal(conmat[keys]))/np.sum(conmat[keys])
    print("Accuracy at SNR=",str(keys)," is ",acc)
    plt.figure()
    plot_confusion_matrix(conmat[keys], labels=labels, title="Confusion Matrix (SNR=%s)"%(str(keys)))
    plt.savefig('Confusion Matrix SNR = %s Acc = %s.jpg'%(str(keys),str(acc)),dpi=300,transparent=True)
```
